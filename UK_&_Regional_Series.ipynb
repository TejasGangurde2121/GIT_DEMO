```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
url="https://www.metoffice.gov.uk/research/climate/maps-and-data/uk-and-regional-series#Yearorder"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
reg=soup.find(id="region")
regions=[d['value'] for d in reg.find_all('option')][1:]
regions
```

```python
para=soup.find(id="parameter")
parameters=[d['value'] for d in para.find_all('option')][1:]
parameters
```

### Finding all regions and all parameter data and storing it

```python
for reg in regions:
    for par in parameters:
        url2=f"https://www.metoffice.gov.uk/pub/data/weather/uk/climate/datasets/{par}/ranked/{reg}.txt"
        page2=requests.get(url2)
        soup2=BeautifulSoup(page2.content,'html5lib')
        s1=soup2.get_text()
        fobj=open(f"uk_reg_climate/{reg}_{par}.txt",'w')
        fobj.write(s1)
        fobj.close()
```
