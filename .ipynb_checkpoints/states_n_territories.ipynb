```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
class States_Territories:
    def __init__(self):
        
        self.url="https://en.wikipedia.org/wiki/States_and_union_territories_of_India"
        self.page=requests.get(self.url)
        self.soup=BeautifulSoup(self.page.content,"html.parser")
        self.table_data=self.soup.find_all(class_= "wikitable sortable plainrowheaders")[0]
        self.table_data_terr=self.soup.find_all(class_= "wikitable sortable plainrowheaders")[1]
        self.data=None
        self.title=None
        self.state=None
        
        self.data1=None
        self.title1=None
        self.terr=None
        
        #print(len(self.table_data))
        
        
    def finding_data(self):
        row=self.table_data.find_all('tr')
        title=[q.get_text().strip() for q in row[0].find_all('th')]
        
        states=[t.find('th').get_text().strip() for t in row[1:]]
        all_data=[f.find_all('td') for f in row[1:]]
        data=[[d.get_text().strip() for d in w] for w in all_data]
        print([len(p) for p in data])
        self.data = data
        self.title=title
        self.state=states
        
    def saving_data(self):
        d={k:[] for k in self.title}
        d['State']=self.state
        for w in self.data:
            if len(w)==10:
                d['ISO'].append(w[0])
                d['Vehiclecode'].append(w[1])
                d['Zone'].append(w[2])
                d['Capital'].append(w[3])
                d['Largest city'].append(w[4])
                d['Statehood'].append(w[5])
                d['Population(2011)[citation needed]'].append(w[6])
                d['Area(km2)'].append(w[7])
                d['Officiallanguages[39]'].append(w[8])
                d['Additional officiallanguages[39]'].append(w[9])
            else:
                d['ISO'].append(w[0])
                d['Vehiclecode'].append(w[1])
                d['Zone'].append(w[2])
                d['Capital'].append(w[3])
                d['Largest city'].append(w[3])
                d['Statehood'].append(w[4])
                d['Population(2011)[citation needed]'].append(w[5])
                d['Area(km2)'].append(w[6])
                d['Officiallanguages[39]'].append(w[7])
                d['Additional officiallanguages[39]'].append(w[8])
                
        df=pd.DataFrame(d)
        df.to_csv("indian_states.csv",index=False)
        return df.head()
    
    def finding_data1(self):
        row=self.table_data_terr.find_all('tr')
        title=[q.get_text().strip() for q in row[0].find_all('th')]
        
        territories=[t.find('th').get_text().strip() for t in row[1:]]
        all_data=[f.find_all('td') for f in row[1:]]
        data=[[d.get_text().strip() for d in w] for w in all_data]
        print([len(p) for p in data])
        self.data1 = data
        self.title1=title
        self.terr=territories
        print(self.title1)
    
    def saving_data_terr(self):
        d={k:[] for k in self.title1}
        d['Union territory']=self.terr
        for w in self.data1:
            if len(w)==10:
                d['ISO 3166-2:IN'].append(w[0])
                d['Vehiclecode'].append(w[1])
                d['Zone'].append(w[2])
                d['Capital'].append(w[3])
                d['Largest city'].append(w[4])
                d['UT established'].append(w[5])
                d['Population'].append(w[6])
                d['Area(km2)'].append(w[7])
                d['Officiallanguages'].append(w[8])
                d['Additional officiallanguages'].append(w[9])
            else:
                d['ISO 3166-2:IN'].append(w[0])
                d['Vehiclecode'].append(w[1])
                d['Zone'].append(w[2])
                d['Capital'].append(w[3])
                d['Largest city'].append(w[3])
                d['UT established'].append(w[4])
                d['Population'].append(w[5])
                d['Area(km2)'].append(w[6])
                d['Officiallanguages'].append(w[7])
                d['Additional officiallanguages'].append(w[8])
                
        df1=pd.DataFrame(d)
        df1.to_csv("indian_territories.csv",index=False)
        return df1.head()
        
        
        
st=   States_Territories()
#st.finding_data()
#st.saving_data()
st.finding_data1()
st.saving_data_terr()
```
