```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
url="https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_1?ie=UTF8&pg=1"
#fetching website
page=requests.get(url)

#passing fetched website
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
books=soup.find_all(class_="p13n-sc-uncoverable-faceout")
books
```

```python
names, rates, revs, prices =[], [], [], []

for first in books:
    try:
        names.append(first.find_all(class_="a-link-normal")[1].get_text())
        prices.append(first.find(class_="p13n-sc-price").get_text())
        rates.append(first.find(class_="a-icon-alt").get_text())
        tp=first.find_all(class_="a-size-small")
        if len(tp)==3:
            revs.append(tp[0].get_text())
        else:
            revs.append(tp[2].get_text())
            
        
    except:
        rates.append(None)
        revs.append(None)
        
len(names), len(rates), len(revs), len(prices)
```

```python
d={'Book Name':names,'Ratings':rates,"No of Reviews":revs, "Price":prices}
d
```

### Amazon second Page


```python
url2="https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_2?ie=UTF8&pg=2"
page2=requests.get(url2)
soup2=BeautifulSoup(page2.content,'html5lib')
books2=soup2.find_all(class_="p13n-sc-uncoverable-faceout")

names, rates, revs, prices =[], [], [], []

for first in books2:
    try:
        names.append(first.find_all(class_="a-link-normal")[1].get_text())
        prices.append(first.find(class_="p13n-sc-price").get_text())
        rates.append(first.find(class_="a-icon-alt").get_text())
        tp=first.find_all(class_="a-size-small")
        if len(tp)==3:
            revs.append(tp[0].get_text())
        else:
            revs.append(tp[2].get_text())
            
        
    except:
        rates.append(None)
        revs.append(None)
        
len(names), len(rates), len(revs), len(prices)
```

```python
d2={'Book Name':names,'Ratings':rates,"No of Reviews":revs, "Price":prices}
d2
```

```python
df2=pd.DataFrame(d2)
df2
```

```python
comb=pd.concat([df,df2])
comb
```
