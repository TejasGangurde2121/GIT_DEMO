```python
from  bs4 import BeautifulSoup 
import requests
import pandas as pd
import re
url="https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"    
page=requests.get(url)
soup=BeautifulSoup (page.content,'html5lib')
soup                
```

```python
nw=soup.find_all(class_="wikitable sortable")[0]
nw
```

```python
mm=nw.find_all("th")
sr=[t.get_text().strip() for t in mm]

sr
```

```python
tg=nw.find_all("td")
data=[l.get_text().strip() for l in tg]
tj=[x for x in data[::6]]
md=[x for x in data[1::6]]
sr=[x for x in data[2::6]]
mga=[x for x in data[3::6]]
tsn=[x for x in data[4::6]]
sjr=[x for x in data[5::6]]

len(tj),len(md),len(sr),len(mga),len(tsn),len(sjr)
```

```python
d1={'Rank':tj,
 'Name':md,
 'Wealth\nChange':sr,
 'Net worth (USD)':mga,
 'Company':tsn,
 'Sources of wealth':sjr}
df=pd.DataFrame(d1)
df
```
